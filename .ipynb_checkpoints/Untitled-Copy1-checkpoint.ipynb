```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.metrics import explained_variance_score, mean_squared_error
%matplotlib inline
```

```python
with open('./Q1_data/data.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
np.random.shuffle(data)
data
```

```python
X = data[:,0]
Y = data[:,1]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
test_X = X[:len(X)//10]
train_X = X[len(X)//10:]
test_Y = Y[:len(X)//10]
train_Y = Y[len(X)//10:]
```

```python
plt.scatter(test_X,test_Y)
plt.show()
```

```python
splits = 10
splittedX = np.array_split(train_X,splits)
splittedY = np.array_split(train_Y,splits)
```

```python
e = list()
for degree in range(1,10):
    error = list()
    variance = list()
    diffs = list()
    all_preds = list()
    for i in range(splits):
        ply = PolynomialFeatures(degree=degree)
        train_X_ = ply.fit_transform(splittedX[i].reshape(-1, 1))
        reg = linear_model.LinearRegression()
        reg.fit(train_X_,splittedY[i])
        
        test_X_ = ply.fit_transform(test_X.reshape(-1,1))
        preds = reg.predict(test_X_)
        #error.append(mean_squared_error(test_Y,preds))
        # error.append(np.mean((preds - test_Y) ** 2))
        all_preds.append(preds)
        variance.append((test_Y-preds))
        diffs.append(preds-test_Y)
    print(np.mean(np.mean(diffs,axis = 0)**2))
    e.append(np.mean(np.mean(diffs,axis = 0)**2))
    #print(np.mean(np.mean((all_preds - np.mean(all_preds,axis = 0))**2,axis = 1)))
# plt.plot(range(1,10),error)
# plt.legend(['MSE'])
# plt.xlabel('Poynomial Degree')
# plt.show()
# plt.plot(range(1,10),variance)
# plt.legend(['Var'])
# plt.xlabel('Poynomial Degree')
# plt.show()
plt.plot(range(1,10),e)
plt.legend(['Bias'])
plt.xlabel('Poynomial Degree')
# plt.show()
# plt.legend(['MSE','Variance','Bias'])
    
```
